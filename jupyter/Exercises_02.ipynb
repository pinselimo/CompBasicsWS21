# Pre-Exercises

Create a program in which the growth of flies is simulated, similar to the VO script. At the beginning there are 10 flies and 10 are added in each time step. Simulate over 20 time steps and plot the result.

```python
# importing matplotlib.pyplot allows us to generate graphs
import matplotlib.pyplot as plt

number_of_flies = 10
timesteps = 20

flies_list = [] # generate an empty list for storing the number of flies
for it in range(timesteps):
    # With append we can add elements to the end of a list
    flies_list.append(number_of_flies)

    # The right hand side (RHS) of the `=` sign is evaluated and *assigned* to the left hand side (LHS)
    number_of_flies = number_of_flies + 10
  
print(flies_list)
plt.plot(flies_list)
plt.title("Number of flies")
plt.xlabel("Timesteps")
plt.ylabel("Flies")
```

Find a clever way to be able to set the growth of "10 flies per time step" at the beginning of the program.

```python
number_of_flies = 10
timesteps = 20
flies_growth = 5

flies_list = list() # an empty list can also be constructed by using `list()`
for it in range(timesteps):
    flies_list.append(number_of_flies)
    # By turning the growth factor into a variable `fliesgrowth` it becomes (*rumble*) variable!
    number_of_flies = number_of_flies + flies_growth
    
print(flies_list)
plt.plot(flies_list)
plt.title("Variable number of flies")
plt.xlabel("Timesteps")
plt.ylabel("N Flies")
```

Add a new command to the very end of the for loop (but still inside the loop). In this line, the value by which the flies grow should be reduced by 1. That means you replace it with a number that is 1 smaller than the value itself.

```python
number_of_flies = 10
flies_growth = 10
timesteps = 20
flies_list = list()
for it in range(timesteps):
    flies_list.append(number_of_flies)
    
    number_of_flies = number_of_flies + flies_growth
    # Now that it is a variable, we can adjust the growth as time progresses
    flies_growth = flies_growth - 1
    
print(flies_list)
plt.plot(flies_list)
plt.title("Number of flies")
plt.xlabel("Time")
plt.ylabel("Flies")
```

## Alternative

During the exercise I showed you an alternative algorithm, I stored it here.

```python
flies = [10]
timesteps = 20

for it in range(timesteps):
    # extract the current number of flies by "indexing" over the list
    number_of_flies = flies[it]
    # With append we can add elements to the end of a list
    flies.append(number_of_flies + 10)
  
print(flies)
plt.plot(flies)
plt.title("Number of flies")
plt.xlabel("Timesteps")
plt.ylabel("Flies")
```

# Exercise

Write a program that simulates a frog population and a fly population at the same time. The starting values are 1 frog and 10 flies. The growth of frogs is initially 1 frog per time step, that of flies 10 flies per time step. After each time step, however, the frog growth increases by 1 and the fly growth decreases by 1. Simulate for 20 steps and plot the result.

```python
number_of_frogs = 1
number_of_flies = 10
growth_frogs = 1
growth_flies = 10

# a variable named with ALL CAPS is considered a *constant* and should never be changed!
TIMESTEPS = 20

list_frogs = list()
list_flies = list()

for it in range(TIMESTEPS):
    list_frogs.append(number_of_frogs)
    list_flies.append(number_of_flies)
    
    number_of_frogs = number_of_frogs + growth_frogs
    growth_frogs = growth_frogs + 1
    
    number_of_flies = number_of_flies + growth_flies
    growth_flies = growth_flies - 1
    
print(list_frogs)
print(list_flies)

plt.plot(list_frogs)
plt.plot(list_flies)
plt.title("Pond Life")
plt.xlabel("Timesteps")
plt.ylabel("Number of animals")
plt.legend(["Frogs", "Flies"])
```
