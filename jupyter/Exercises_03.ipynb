Create a program by growing a rabbit population exponentially, similar to the VO script. The starting population is 100, the growth is 20 percent. Simulate 20 time steps.

```python
# importing matplotlib.pyplot allows us to generate graphs
import matplotlib.pyplot as plt

rabbit_population = 100
RABBIT_GROWTH = 0.2 # 20 %
TIMESTEPS = 20

rabbit_recorder = [rabbit_population]
for it in range(TIMESTEPS):
    rabbits_born = rabbit_population * RABBIT_GROWTH
    rabbit_population = rabbit_population + rabbits_born
    
    rabbit_recorder.append(
            rabbit_population
    )
    
print(rabbit_recorder)
plt.plot(rabbit_recorder)
plt.title("Rabbit population chart")
plt.xlabel("Timesteps")
plt.ylabel("Population")
```

Rewrite the if-statement from the last preliminary exercise: Not the current number of rabbits should be decisive, but the number of rabbits in the previous step. For this you will need a list.

```python
rabbit_population = 100
RABBIT_GROWTH = 0.2 # 20 %
TIMESTEPS = 20

rabbit_recorder = [rabbit_population]
for steps in range(TIMESTEPS):
    if steps >= 5:
        # one - off errors are sometimes hard to track down
        # both, indexing starting with 0 and the time we append
        # to the list (before or after our operations) can alter
        # just where in the list we find the _correct_ number
        # of rabbits we want to find.
        # For now you can use the "print" function to check
        # if your assumptions are met.
        prev_rabbit_population = rabbit_recorder[ steps-1 ]
        if prev_rabbit_population < 200:
            rabbits_born = rabbit_population * RABBIT_GROWTH
            rabbit_population = rabbit_population + rabbits_born
    rabbit_recorder.append(rabbit_population)
    
print(rabbit_recorder)
plt.plot(rabbit_recorder)
plt.title("Rabbit population chart")
plt.xlabel("Timesteps")
plt.ylabel("Population")
```

# Exercise

Write a program in which you simulate the growth of a rabbit population. At the beginning there are 100 rabbits and in each time step 5% are added. After time step 10, however, the growth behavior should be different: Whenever there have been more than 1000 rabbits 5 steps before the current time step, half of the rabbits migrate to another habitat (and are thus removed from this system). In this case (instead of growth) the number of rabbits should be halved. Simulate 100 time steps.

```python
rabbit_pop = 100
GROWTH = 0.05 # 5 %
TIMESTEPS = 100
# We can pull out some values and store them in CONSTANT variables.
# That way we can store those *parameters* in one place
# and easily change them throughout the programm quickly.
LIMIT_10 = 10
LOOK_BACK = 5
LIMIT_RABBITS = 1000

recorder = [rabbit_pop]
for step in range(TIMESTEPS):
    rabbits_born = rabbit_pop * GROWTH
    
    if step >= LIMIT_10:
        rabbits_past = recorder[ step - LOOK_BACK ]
        if rabbits_past > LIMIT_RABBITS:
            rabbit_pop = rabbit_pop / 2
        else: # only if the limit was not reached we want the population to grow
            rabbit_pop = rabbit_pop + rabbits_born
    else: # in case the temporal limit was not met yet there's just normal growth
        rabbit_pop = rabbit_pop + rabbits_born
    
    recorder.append(rabbit_pop)
    
print(recorder)
plt.plot(recorder)
plt.title("Big rabbit population chart")
plt.xlabel("Timesteps")
plt.ylabel("Rabbits")
```
